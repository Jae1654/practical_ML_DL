### dot and element-wise operation


```python
A = np.arange(2*2).reshape(2,2)
B = np.array([[3,4],
              [5,6]])

print(f'Mat A:{A}')
print(f'Mat A:{B}')
print("===========")

# dot product AB
print("Dot Product")
print(A @ B)
print(np.dot(A,B))
print("===========")

print("Element-wise operation")
print(A * B)
print(np.multiply(A,B))
print("===========")
```
